### **2. Short Derivation (6 Marks)**
Derive the expression for the inverse of a 2×2 matrix:

$$
A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}
\quad \text{such that} \quad 
A^{-1} = \frac{1}{ad-bc}\begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
$$

Then, explain under what condition a matrix is **not invertible**.


$$AA^{-1} = A^{-1}A = \begin{bmatrix} 1 & 0 \\ 0 & 1  \end{bmatrix}$$

$$\quad \text{Let} \quad  A^{-1} = \begin{bmatrix} p & q \\ r & s \end{bmatrix}$$

$$
\quad \text{Then,} \quad AA^{-1} = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \begin{bmatrix} p & q \\ r & s  \end{bmatrix} = \begin{bmatrix} 1 & 0 \\ 0 & 1  \end{bmatrix}
$$

$$
AA^{-1} = \begin{bmatrix} ap+br & aq+bs \\ cp+dr & cq+ds \end{bmatrix} = \begin{bmatrix} 1 & 0 \\ 0 & 1  \end{bmatrix}
$$

$$ \quad \text{Therefore,} \quad\quad\quad\quad\quad\quad\quad\quad $$
$$ap+br = 1 \quad equ(1); \quad cp+dr = 0 \quad equ(2);$$
$$aq+bs = 0 \quad equ(3); \quad cq+ds  = 1 \quad equ(4);$$
$$ \quad \text{writing equ(1) and equ(2) in terms of p and equating them}: $$
$$\frac{1 - br}{a} = \frac{-dr}{c}$$
$$c - brc = -adr$$
$$adr - brc = -c$$
$$\boxed{r = \frac{-c}{ad - bc}}$$
$$ \quad \text{writing equ(1) and equ(2) in terms of r and equating them}: $$
$$\frac{1 - ap}{b} = \frac{-cp}{d}$$
$$-bcp = d - adp$$
$$\boxed{p = \frac{d}{ad - bc}}$$
$$ \quad \text{writing equ(3) and equ(4) in terms of q and equating them}: $$
$$\frac{-bs}{a} = \frac{1 - ds}{c}$$
$$-bcs = a - ads$$
$$\boxed{s = \frac{a}{ad - bc}}$$
$$ \quad \text{writing equ(3) and equ(4) in terms of s and equating them}: $$
$$\frac{-aq}{b} = \frac{1 - cq}{d}$$
$$-adq = b - bcq$$
$$-b = adq - bcq$$
$$\boxed{q = \frac{-b}{ad - bc}}$$

$$
\text{Thus,} \quad A^{-1} = \begin{bmatrix} p & q \\ r & s \end{bmatrix} = \begin{bmatrix} \frac{d}{ad - bc} & \frac{-b}{ad - bc} \\ \frac{-c}{ad - bc} & \frac{a}{ad - bc}  \end{bmatrix} = \frac{1}{ad - bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
$$

$$
A= \begin{bmatrix} a & b \\ c & d \end{bmatrix} \quad \text{and} \quad A^{-1} = \frac{1}{ad - bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
$$

$$
\text{The expression, } \boxed{ad - bc} \quad \text{is known as the determinant}
$$

$$
A^{-1} \quad \text{is defined, that is, A is invertible if and only if} \quad \boxed{ad - bc \neq 0}
$$

$$
\text{Therefore, we can conclude that a square matrix is invertible provided that its determinant is not equal to zero. Otherwise, it is not invertible} \quad

### **3. Matrix Arithmetic (6 Marks)**

Given

$$
A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix},
\quad 
B = \begin{bmatrix} 9 & 8 & 7 \\ 6 & 5 & 4 \\ 3 & 2 & 1 \end{bmatrix}
$$

a) Compute \(A + B\) and \(2A - 3B\) using **pure Python**.  
b) Repeat the same using **NumPy** and compare the results.


```python
import numpy as np
from pprint import pp

    
def sum_2d(P, Q):
    result = []
    for p, q in zip(P, Q):
        inner_list = []
        for x, y in zip(p, q):
            inner_list.append(x + y)
        result.append(inner_list)
    return result

def scale_2d(P, scaler):
    result = []
    for p in P:
        new_p = []
        for x in p:
            new_p.append(x * scaler)
        result.append(new_p)
    return result

A = [[1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]]
B = [[9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]]

ab_sum_py = sum_2d(A, B)
scaled_sum_py = sum_2d(scale_2d(A, 2), scale_2d(B, -3))
print('sum in pure python: ')
pp(np.array(ab_sum_py))
print('scaled sum in pure python: ')
pp(np.array(scaled_sum_py))
print('\n')

arr_A = np.array(A)
arr_B = np.array(B)
ab_sum_numpy = arr_A + arr_B
scaled_sum_numpy = 2*arr_A - 3*arr_B
print('sum using numpy: ')
pp(ab_sum_numpy)
print('scaled sum using numpy: ')
pp(scaled_sum_numpy)
print('\n')

print(f'Sum in pure python equals sum in numpy: {np.array_equal(ab_sum_py, ab_sum_numpy)}')
print(f'Scaled sum in pure python equals scaled sum in numpy: {np.array_equal(scaled_sum_py, scaled_sum_numpy)}')
```

### **4. Transpose and Symmetry (4 Marks)**

a) Write a pure Python function `is_symmetric(matrix)` that checks whether a given matrix is symmetric.  
b) Test your function on

$$
S = \begin{bmatrix} 2 & 1 & 3 \\ 1 & 5 & 0 \\ 3 & 0 & 6 \end{bmatrix}
$$

and visualize both $(S)$ and $(S^T)$ using `matplotlib`.


```python
def is_symmetric(matrix) -> bool:
    for row in range(len(matrix)):
        for col in range(len(matrix[row])):
            if matrix[row][col] != matrix[col][row]:
                return False
    return True

S = [[2, 1, 3],
    [1, 5, 0],
    [3, 0, 0]]

print(f'Matrix S is symmetric: {is_symmetric(S)}')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

S = np.array(S)

S_T = S.T 

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
X1, Y1 = np.meshgrid(range(S.shape[1]), range(S.shape[0]))
surf1 = ax1.plot_surface(X1, Y1, S, cmap='Blues', alpha=0.9, 
                         edgecolor='darkblue', linewidth=0.5, vmin=0, vmax=5)
ax1.set_xlabel('X', fontsize=11, fontweight='bold')
ax1.set_ylabel('Y', fontsize=11, fontweight='bold')
ax1.set_zlabel('Z', fontsize=11, fontweight='bold')
ax1.set_title('Original Matrix S', fontsize=13, fontweight='bold', pad=20, color='darkblue')
ax1.view_init(elev=25, azim=45)
cbar1 = fig.colorbar(surf1, ax=ax1, shrink=0.6, pad=0.1)
cbar1.set_label('Value', fontsize=10)


ax2 = fig.add_subplot(122, projection='3d')
X2, Y2 = np.meshgrid(range(S_T.shape[1]), range(S_T.shape[0]))
surf2 = ax2.plot_surface(X2, Y2, S_T, cmap='Oranges', alpha=0.9, 
                         edgecolor='darkorange', linewidth=0.5, vmin=0, vmax=5)
ax2.set_xlabel('X', fontsize=11, fontweight='bold')
ax2.set_ylabel('Y', fontsize=11, fontweight='bold')
ax2.set_zlabel('Z', fontsize=11, fontweight='bold')
ax2.set_title('Transpose of S (S^T)', fontsize=13, fontweight='bold', pad=20, color='darkorange')
ax2.view_init(elev=25, azim=45)
cbar2 = fig.colorbar(surf2, ax=ax2, shrink=0.6, pad=0.1)
cbar2.set_label('Value', fontsize=10)

plt.suptitle('Matrix S and Its Transpose', 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

### **5. Solving Linear Systems (5 Marks)**

Consider the system

$$
\begin{cases}
2x_1 + 3x_2 = 8 \\
x_1 + 2x_2 = 5
\end{cases}
$$

a) Solve it manually using the inverse formula.  
b) Verify your result with NumPy’s `linalg.solve()` function.


```python
# Your code here
coef = np.array([[2, 3],
                [1, 2]])
intercept = np.array([8, 5])

inv_coef = np.linalg.inv(coef)
solution = inv_coef@intercept

numpy_solution = np.linalg.solve(coef, intercept)
print(f'Pure Python solution is equal to Numpy solution: {np.array_equal(numpy_solution,solution)}')
```

### **6. Rank and Null Space (5 Marks)**

Let

$$
A = \begin{bmatrix} 1 & 2 & 3 \\ 2 & 4 & 6 \\ 1 & 1 & 1 \end{bmatrix}
$$

a) Compute its rank manually using row reduction (you may use Python loops).  
b) Find the null space using NumPy (`np.linalg.svd`) and interpret what the null space tells you about the columns of \(A\).


```python
# Your code here
A = [[1, 2, 3],
    [2, 4, 6],
    [1, 1, 1]]

def rank_finder(matrix):
    mat = np.array([row.copy() for row in matrix], dtype=float)
    rank = 0
    rows = len(mat)
    cols = len(mat[0])

    for row in range(min(rows, cols)):
        current_pivot_element = mat[row][row]
        if abs(current_pivot_element) < 1e-10:
            for array in range(row, rows):
                if abs(mat[array][row]) > 1e-10:
                    mat[row], mat[array] = mat[array], mat[row]
                    current_pivot_element = mat[row][row]
                    break
                
        if abs(current_pivot_element) < 1e-10:
            continue
            
        for others in range(row+1, rows):
            lever = mat[others][row]/current_pivot_element
            mat[others] -= lever*mat[row]
        mat[row] /= current_pivot_element
    
    for element in range(min(rows, cols)):
        if abs(mat[element][element]) > 1e-10:
            rank += 1
    
    return rank

(rank_finder(A))
```

### **7. Linear Regression with Normal Equation (10 Marks)**

You are given a small dataset:

| Hours Studied (x) | Exam Score (y) |
|--------------------|----------------|
| 1 | 2 |
| 2 | 4 |
| 3 | 6 |
| 4 | 8 |

a) Represent this dataset as a matrix $( X = [x_i, x_j] )$ and vector $( y )$.  
b) Compute the optimal weights $( w^* = (X^TX)^{-1}X^Ty )$ using either pure Python or NumPy.  
c) Plot the fitted line against the data points using `matplotlib`.  
d) Explain briefly how this relates to what happens in linear regression training in Machine Learning.


```python
# Your code here
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])

w = np.linalg.inv(X.T@X)@X.T@y
print('Weights, w: \n')
pp(w)
```

```python
predictions = w@X.T

fig, ax = plt.subplots(figsize=(5, 4))

ax.scatter(X, y, color='red', label='data labels')

ax.plot(X, predictions, label='model predictions')

plt.xlabel('X - values')
plt.ylabel('y - values')
plt.title('Graph of Feature X againt target y')
ax.legend(fontsize=10)
ax.grid(True)
plt.show()
```

During training, a linear regression model seeks to find the optimal set of weights for each feature in the dataset. 

The optimal set of weight is that which reduces the distance between the data points from the dataset (represented by the red points in plot above) and the model predictions (represented by the blue line above). 

The straight line represents the models predictions and the red dots represent the data points (the observations) from the dataset. The goal of a linear regression model during training in machine learning is to find a line that best matches the data point just as the blue line above.
