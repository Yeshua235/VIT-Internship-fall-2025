#### 1.1  EUCLIDEAN NORMS (L2 NORMS)

The **Euclidean norm** of a vector is defined as:

$$
\| \mathbf{v} \|_2 = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}
$$

<br>

üí¨ Intuition: <br>
Like using a ruler to measure the straight-line length of a vector in space.

````python
# Function to compute L2 norm manually
def l2_norm(v):
    total = 0
    for x in v:
        total += x**2
    return total**0.5

# Example
v = [3, 4]
print("Vector:", v)
print("L2 Norm =", l2_norm(v))
````

````python
# OR
import math 
def l2_norm(vec):
    """L2 (Euclidean) norm: sqrt(sum of squares)"""
    s = 0.0
    for v in vec:
        s += v * v  
    return math.sqrt(s)


# Example
v = [3, 4]
print("Vector:", v)
print("L2 Norm =", l2_norm(v))
````

````python
# using numpy
import numpy as np

v = np.array([3, 4])
l2 = np.linalg.norm(v)
print("L2 Norm =", l2)
````

#### 1.2 Manhattan Norm (L1 Norm)

The **Manhattan norm** of a vector is defined as:

$$
\| \mathbf{v} \|_1 = |v_1| + |v_2| + \cdots + |v_n|
$$

üí¨ **Intuition:**  
This measures the **‚Äúcity block‚Äù distance** ‚Äî how far you would walk **along gridlines** (not diagonally).

````python
# Function to compute L1 norm manually
def l1_norm(v):
    total = 0
    for x in v:
        total += abs(x)
        # total = total + abs(x)
    return total

v = [3, -4]
print("Vector:", v)
print("L1 Norm =", l1_norm(v))
````

````python
#using numpy
import numpy as np

v = np.array([3, -4])
l1 = np.sum(np.abs(v))
# print("Vector:", v)
print("L1 Norm =", l1)
````

#### 1.3 Distance Between Two Points

Given two vectors $\mathbf{a}$ and $\mathbf{b}$, the **Euclidean distance** between them is:

$$
d_{L_2}(\mathbf{a}, \mathbf{b}) 
= \sqrt{(a_1 - b_1)^2 + (a_2 - b_2)^2 + \cdots + (a_n - b_n)^2}
$$

````python
def euclidean_distance(a, b):
    total = 0
    for i in range(len(a)):
        total += (a[i] - b[i])**2
    return total**0.5

# a[0] = 1, a[1] = 2
a = [1, 2]
b = [4, 6]
print("Distance between", a, "and", b, "=", euclidean_distance(a, b))
````

````python
import numpy as np

a = np.array([1, 2])
b = np.array([4, 6])
dist = np.linalg.norm(a - b)
print("Euclidean Distance =", dist)
````

#### 1.4 Manhattan Distance (L‚ÇÅ Distance)

Given two vectors $\mathbf{a}$ and $\mathbf{b}$, the **Manhattan distance** between them is:

$$
d_{L_1}(\mathbf{a}, \mathbf{b}) 
= |a_1 - b_1| + |a_2 - b_2| + \cdots + |a_n - b_n|
$$

Fun fact:
Also called **‚ÄúTaxicab‚Äù or ‚ÄúCity Block‚Äù distance** ‚Äî like walking along streets instead of diagonally.

````python
def manhattan_distance(a, b):
    total = 0
    for i in range(len(a)):
        total += abs(a[i] - b[i])
    return total
# a[1] = 2
a = [1, 2]
b = [4, 6]
print("Manhattan distance between", a, "and", b, "=", manhattan_distance(a, b))
````

````python
# using numpy
import numpy as np

def manhattan_distance(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.sum(np.abs(a - b))

a = [1, 2]
b = [4, 6]
print("Manhattan distance between", a, "and", b, "=", manhattan_distance(a, b))
````

##### üî¢ Dot Product (Inner Product)

The **dot product** of two vectors in **n-dimensional space** is defined as:

$$
\mathbf{v} \cdot \mathbf{w} = \sum_{i=1}^{n} v_i w_i
$$

**Example for 2D vectors:**

$$
\begin{bmatrix} v_1 \\ v_2 \end{bmatrix}
\cdot
\begin{bmatrix} w_1 \\ w_2 \end{bmatrix}
= v_1 w_1 + v_2 w_2
$$

**Example for 3D vectors:**

$$
\begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix}
\cdot
\begin{bmatrix} w_1 \\ w_2 \\ w_3 \end{bmatrix}
= v_1 w_1 + v_2 w_2 + v_3 w_3
$$

#### Did you know??

Let a vector be  
$$\mathbf{v} = (v_1, v_2, \dots, v_n)$$

The **dot product** of a vector with itself is:  
$$\mathbf{v} \cdot \mathbf{v} = v_1^2 + v_2^2 + \dots + v_n^2$$

The **Euclidean norm** (length) of the vector is:  
$$\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}$$

‚úÖ Therefore:  
$$\mathbf{v} \cdot \mathbf{v} = \|\mathbf{v}\|^2$$

So, the dot product of a vector with itself equals the **square of its Euclidean norm**, not the norm itself.

**Example:**  
$$\mathbf{v} = (3, 4)$$  
$$\mathbf{v} \cdot \mathbf{v} = 3^2 + 4^2 = 25, \quad \|\mathbf{v}\| = 5$$

### üìò Inner Product Axioms

Let $V$ be a vector space over $\mathbb{R}$.  
The inner product $\langle \mathbf{u}, \mathbf{v} \rangle$ must satisfy **four properties (axioms):**



---

#### üß© Axiom 1 ‚Äî Symmetry

$$
\langle \mathbf{u}, \mathbf{v} \rangle = \langle \mathbf{v}, \mathbf{u} \rangle
$$

üí¨ **Meaning:**  
The order of the vectors doesn‚Äôt matter ‚Äî swapping them does **not** change the result.


````python
def dot_product(v, w):
    return sum(v[i] * w[i] for i in range(len(v)))

u = [2, 3]
v = [4, 1]

lhs = dot_product(u, v)
rhs = dot_product(v, u)
print("Symmetry holds:", lhs == rhs)
````

````python
# using numpy
import numpy as np

u = np.array([2, 3])
v = np.array([4, 1])

print("Symmetry holds:", np.dot(u, v) == np.dot(v, u))
````

### üß© Axiom 2 ‚Äî Linearity in the First Argument

$$
\langle a\mathbf{u} + b\mathbf{v}, \mathbf{w} \rangle
= a \langle \mathbf{u}, \mathbf{w} \rangle + b \langle \mathbf{v}, \mathbf{w} \rangle
$$


üí¨ **Meaning:**  
Scalars and addition can be distributed inside the **first argument** ‚Äî the inner product is **linear** in the first vector.

````python
# dot product function defined above

a, b = 2, -1
u = [1, 2]
v = [3, 1]
w = [0, 4]

lhs = dot_product([a*u[i] + b*v[i] for i in range(len(u))], w)
rhs = a * dot_product(u, w) + b * dot_product(v, w)

print("Linearity in first argument holds:", abs(lhs - rhs) < 1e-9)
````

````python
# using numpy
u = np.array([1, 2])
v = np.array([3, 1])
w = np.array([0, 4])
a, b = 2, -1

lhs = np.dot(a*u + b*v, w)
rhs = a*np.dot(u, w) + b*np.dot(v, w)
print("Linearity holds:", np.allclose(lhs, rhs))
````

### üß© Axiom 3a ‚Äî Positivity

$$
\langle \mathbf{v}, \mathbf{v} \rangle \ge 0
$$

üí¨ **Meaning:**  
The inner product of any vector with itself is always non-negative ‚Äî it represents squared length.

````python
v = [3, -4]
inner = dot_product(v, v)
print("‚ü®v,v‚ü© =", inner, "‚Üí Always ‚â• 0:", inner >= 0)
````

### üß© Axiom 3b ‚Äî Positive Definiteness

$$
\langle \mathbf{v}, \mathbf{v} \rangle = 0 \;\;\Longleftrightarrow\;\; \mathbf{v} = \mathbf{0}
$$

üí¨ **Meaning:**  
Only the **zero vector** has zero length.

````python
v = [0, 0]
inner = dot_product(v, v)
print("Inner product =", inner)
print("Vector is zero vector:", all(x == 0 for x in v))
````

### üåü Weighted Inner Product

Some inner products give **more importance (weight)** to certain components:

$$
\langle \mathbf{u}, \mathbf{v} \rangle_W 
= \sum_{i=1}^{n} w_i \, u_i \, v_i 
\qquad \text{where } w_i > 0
$$

üí° **Meaning:** Each component is scaled by a **positive weight** $ w_i \$. This is useful when some features are considered **more important** than others.

````python
# Weighted inner product
def weighted_dot(u, v, w):
    total = 0
    for x, y, wi in zip(u, v, w):
        total += wi * x * y
    return total

# Example
u = [1, 2]
v = [3, 4]
w = [2, 0.5]  # weights: first feature more important
print("Weighted Inner Product =", weighted_dot(u, v, w))
````

### üß† Concept: Unit Vector

A **unit vector** is simply a vector whose **length (norm) is equal to 1**.

If $\mathbf{v}$ is any non-zero vector, its corresponding unit vector $\hat{\mathbf{v}}$ is given by:

$$
\hat{\mathbf{v}} = \frac{\mathbf{v}}{\|\mathbf{v}\|}
$$

üí¨ **In simple words:**  
Divide **every component** of the vector by its **norm (length)** to get a new vector that points in the **same direction**, but has **length = 1**.


---

### ‚úÖ Example

Let $\mathbf{v} = [3, 4]$.

First, compute its L‚ÇÇ norm:

$$
\|\mathbf{v}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = 5
$$

Now, the **unit vector** is:

$$
\hat{\mathbf{v}} = \frac{1}{5} [3, 4] = \left[ \frac{3}{5}, \frac{4}{5} \right]
$$

So the direction is **unchanged**, but now its **length = 1** ‚úÖ

````python
def l2_norm(v):
    total = 0
    for x in v:
        total += x**2
    return total ** 0.5

def unit_vector(v):
    norm = l2_norm(v)
    if norm == 0:
        raise ValueError("Zero vector has no direction, cannot normalize.")
    return [x / norm for x in v]

# Example
v = [3, 4]
u = unit_vector(v)

print("Vector v:", v)
print("Unit vector √ª:", u)

# Confirm it truly has length 1
print("Length of √ª =", l2_norm(u))
````

````python
# using numpy
import numpy as np

v = np.array([3, 4])
unit_v = v / np.linalg.norm(v)

print("Unit vector =", unit_v)
print("Length =", np.linalg.norm(unit_v))
````

### SECTION 3: INEQUALITIES

### üß† Starting Point ‚Äî Geometric Definition of Inner Product

The **inner product (dot product)** between two vectors is defined as:

$$
\mathbf{v} \cdot \mathbf{w}
= \|\mathbf{v}\| \, \|\mathbf{w}\| \cos(\theta)
$$

where:

- $\theta$ is the angle between $\mathbf{v}$ and $\mathbf{w}$
- $\|\mathbf{v}\|$ and $\|\mathbf{w}\|$ are their magnitudes (norms)

üí¨ **Meaning:**  
The dot product measures **how much one vector points in the direction of the other**.

---
### üå∏ From the Formula to the Inequality

We know that for any real angle $\theta$:

$$
-1 \le \cos(\theta) \le 1
$$

Substitute this into the geometric definition of the inner product:

$$
\mathbf{v} \cdot \mathbf{w}
= \|\mathbf{v}\| \, \|\mathbf{w}\| \cos(\theta)
$$

Taking the absolute value of both sides:

$$
|\mathbf{v} \cdot \mathbf{w}|
= \|\mathbf{v}\| \, \|\mathbf{w}\| \, |\cos(\theta)|
$$

Since $|\cos(\theta)| \le 1$, it follows that:

$$
|\mathbf{v} \cdot \mathbf{w}| \le \|\mathbf{v}\| \, \|\mathbf{w}\|
$$

üí• **This is exactly the Cauchy‚ÄìSchwarz Inequality!**


````python
# l2_norm defined earlier!

v = [2, 3]
w = [4, 1]

lhs = abs(dot_product(v, w))
rhs = l2_norm(v) * l2_norm(w)

print(f"|v¬∑w| = {lhs:.4f}")
print(f"||v||¬∑||w|| = {rhs:.4f}")
print("Cauchy‚ÄìSchwarz holds:", lhs <= rhs)
````

### üß© Equality Condition

Equality in the Cauchy‚ÄìSchwarz Inequality happens **only when**  
$\mathbf{v}$ and $\mathbf{w}$ are **linearly dependent**, i.e., one is a scalar multiple of the other. That is the vectors are parallel!

üí¨ Example:

If  
$$
\mathbf{w} = k \mathbf{v}
$$  
then  
$$
|\mathbf{v} \cdot \mathbf{w}| = \|\mathbf{v}\| \, \|\mathbf{w}\|
$$

````python
v = [1, 2]
w = [2, 4]  # w = 2v

lhs = abs(dot_product(v, w))
rhs = l2_norm(v) * l2_norm(w)

print(f"|v¬∑w| = {lhs:.4f}")
print(f"||v||¬∑||w|| = {rhs:.4f}")
print("Cauchy‚ÄìSchwarz equality holds:", abs(lhs - rhs) < 1e-9)
````

### üåø From Cauchy‚ÄìSchwarz to Triangle Inequality

Now that we know

$$
|\mathbf{v} \cdot \mathbf{w}| \le \|\mathbf{v}\| \, \|\mathbf{w}\|
$$

we can use this to prove that the **triangle inequality** holds for norms.

Start with:

$$
\|\mathbf{v} + \mathbf{w}\|^2 
= (\mathbf{v} + \mathbf{w}) \cdot (\mathbf{v} + \mathbf{w}) 
= \|\mathbf{v}\|^2 + 2(\mathbf{v} \cdot \mathbf{w}) + \|\mathbf{w}\|^2
$$

Using Cauchy‚ÄìSchwarz:

$$
\mathbf{v} \cdot \mathbf{w} \le \|\mathbf{v}\| \, \|\mathbf{w}\|
$$

So:

$$
\|\mathbf{v} + \mathbf{w}\|^2 
\le (\|\mathbf{v}\| + \|\mathbf{w}\|)^2
$$

Taking the square root of both sides gives:

$$
\|\mathbf{v} + \mathbf{w}\| \le \|\mathbf{v}\| + \|\mathbf{w}\|
$$

üéâ **That‚Äôs the Triangle Inequality!**

````python
def verify_triangle(v, w):
    lhs = l2_norm([v[i] + w[i] for i in range(len(v))])
    rhs = l2_norm(v) + l2_norm(w)
    return lhs <= rhs, lhs, rhs

v = [3, 0]
w = [0, 4]

holds, lhs, rhs = verify_triangle(v, w)
print("Triangle inequality holds:", holds)
print("||v+w|| =", lhs, "‚â§ ||v|| + ||w|| =", rhs)
````

### Soooo what I'm I saying? 

The geometric formula

$$
\mathbf{v} \cdot \mathbf{w} = \|\mathbf{v}\| \, \|\mathbf{w}\| \cos \theta
$$

gives rise to the **Cauchy‚ÄìSchwarz inequality** because cosine is always between **‚àí1 and 1**.

And once we know that, we can prove the **Triangle Inequality** ‚Äî the mathematical version of  
‚Äúthe shortest path between two points is a straight line.‚Äù

#### üß† 1. Orthogonality (Perpendicularity)

Two vectors $\mathbf{v}$ and $\mathbf{w}$ are said to be **orthogonal** if **their inner product (dot product)** is zero:

$$
\langle \mathbf{v}, \mathbf{w} \rangle = 0
$$

üí¨ **Meaning:**  
They are at right angles (90¬∞ apart), so they share no ‚Äúdirectional overlap" or they have no influence on each other.  
In **ML**, orthogonal vectors represent **independent features** ‚Äî no correlation!


##### üîç Example from our lecture slide
---

Given:

$$
\mathbf{v} = [1,\, 2], 
\quad 
\mathbf{w} = [6,\,-3]
$$

Compute the dot product:

$$
\mathbf{v} \cdot \mathbf{w}
= 1 \times 6 + 2 \times (-3)
= 6 - 6 = 0
$$

‚úÖ **Yes ‚Äî they are orthogonal!**


````python
# function already defined above
def is_orthogonal(v, w):
    return dot_product(v, w) == 0

v = [1, 0]
w = [0, 5]

print("Are v and w orthogonal?", is_orthogonal(v, w))
````

### üß≠ Orthogonal Basis

A **basis** is a set of vectors that can **span an entire vector space**.

An **orthogonal basis** is a basis where **every pair of vectors is orthogonal** to each other. For example:

$$
\mathbf{u}_1 \cdot \mathbf{u}_2 = 0, \quad
\mathbf{u}_1 \cdot \mathbf{u}_3 = 0, \quad
\mathbf{u}_2 \cdot \mathbf{u}_3 = 0, \ \text{etc.}
$$

‚úÖ **Why this is powerful:**

- Every orthogonal set of **non-zero vectors is automatically linearly independent**  
- **Coordinates become super easy to compute** (often just dot products)  
- If you also **normalize each vector**, you get an **orthonormal basis** (same direction, but each has length = 1)  


````python
import numpy as np

# Function to check if a set of vectors form an orthogonal basis
def is_orthogonal_basis(vectors):
    for i in range(len(vectors)):
        for j in range(i + 1, len(vectors)):  # only compare different pairs
            if np.dot(vectors[i], vectors[j]) != 0:
                return False
    return True

# Example
u1 = np.array([1, 0])
u2 = np.array([0, 1])
print("Is orthogonal basis?", is_orthogonal_basis([u1, u2]))
````

### Orthonormal Basis

An **orthonormal basis** is simply an **orthogonal** basis where every vector has **length 1**.

- **Orthogonal** ‚Üí angle = $90^\circ$ ‚Üí $u_i \cdot u_j = 0$  
- **Orthonormal** ‚Üí angle = $90^\circ$ **and** each vector has **unit length**

$$\|u_i\| = 1 \quad \text{and} \quad \langle u_i, u_j \rangle = 0 \quad \text{for } i \ne j$$

---

‚úÖ **Why is an orthonormal basis even better?**

Because calculations become **super simple**:

- The **dot product becomes the exact coordinate**

If $\{u_1, u_2\}$ is orthonormal, then

$$c_1 = \langle u_1, v \rangle$$

(no need to divide by $\|u_i\|^2$ like in a normal orthogonal basis)

---

**Widely used in**: machine learning, signal processing, PCA, Fourier basis, word2vec, etc.

**Makes** projection, decomposition, and distance computations **much faster and numerically stable.**


````python
def is_orthonormal_basis(vectors):
    for i in range(len(vectors)):
        for j in range(len(vectors)):
            dot = np.dot(vectors[i], vectors[j])
            if i == j and dot != 1:   # length must be 1
                return False
            if i != j and dot != 0:   # others must be orthogonal
                return False
    return True

# Test sample
u1 = np.array([1/np.sqrt(2), 1/np.sqrt(2)])
u2 = np.array([1/np.sqrt(2), -1/np.sqrt(2)])
print(is_orthonormal_basis([u1, u2]))
````

##### How the function checks for an orthonormal basis

It goes through each vector **one by one**,  
then compares it with **every other vector ‚Äî including itself**.

It performs **two tests**:

| **Case** | **What it checks** | **Expected dot product** |
|----------|---------------------|---------------------------|
| Same vector ($i = j$) | Is its length = 1? | $\,\langle v, v \rangle = 1$ |
| Different vectors ($i \ne j$) | Are they orthogonal? | $\,\langle v, w \rangle = 0$ |

---

‚úÖ If **all conditions are satisfied**, the function returns **`True`**  
‚ùå If **any condition fails**, it immediately returns **`False`**
