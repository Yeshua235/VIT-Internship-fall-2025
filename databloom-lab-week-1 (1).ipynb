# SECTION 1: **Vector Addition**

Example 1

We define two vectors:

$$
\mathbf{v} =
\begin{pmatrix}
2 \\
3
\end{pmatrix},
\quad
\mathbf{w} =
\begin{pmatrix}
2 \\
1
\end{pmatrix}
$$

We can write this as:

$$
\mathbf{v} = [2, 3]^T, \quad \mathbf{w} = [2, 1]^T
$$

To add them, we add corresponding components:

$$
\mathbf{v} + \mathbf{w} =
\begin{pmatrix}
v_1 + w_1 \\
v_2 + w_2
\end{pmatrix}
=
\begin{pmatrix}
2 + 2 \\
3 + 1
\end{pmatrix}
=
\begin{pmatrix}
4 \\
4
\end{pmatrix}
$$

Therefore:

$$
\mathbf{v} + \mathbf{w} = [4, 4]^T
$$


````python
# vectors as list
v = [2, 3]
w = [2, 1]

# element-wise addition
z = [vi + wi for vi, wi in zip(v, w)]
print('v + w =', z)
````

````python
# OR
# Two 2D vectors
v = [2, 3]
w = [2, 1]

# Manual vector addition
z = [v[i] + w[i] for i in range(len(v))]
print("v + w =", z)
````

````python
v = [2, 7]
w = [2, 4]

# Manual vector addition
z = [v[i] + w[i] for i in range(len(v))]
print("v + w =", z)
````

````python
v = [2, 7, 7, 9]
w = [2, 4, 8, 9]

# Manual vector addition
z = [v[i] + w[i] for i in range(len(v))]
print("v + w =", z)
````

Note:

zip(v, w) pairs up elements from both lists
[(2, 2), (3, 1)]
For each pair (vi, wi), it adds them together:

1st pair: 2 + 2 = 4

2nd pair: 3 + 1 = 4

So the resulting list is [4, 4]

# SECTION 4: Linear Combinations, Span, and Linear Independence

A **linear combination** of vectors is formed by scaling some vectors and adding them together.

If you have vectors

$$
\vec{v}_1, \; \vec{v}_2, \; \dots, \; \vec{v}_k
$$

and scalars

$$
c_1, \; c_2, \; \dots, \; c_k,
$$

then:

**Linear Combination:**

$$
\vec{x} = c_1 \vec{v}_1 + c_2 \vec{v}_2 + \cdots + c_k \vec{v}_k
$$

---

üí¨ **In words:**  
You can reach new points in space by *stretching* and *adding* existing vectors.  
This idea is the backbone of ML ‚Äî your model‚Äôs predictions are literally **linear combinations of weights and inputs**!

keep **weights** in mind later we will learn about them!


<br>

---
#### Example
<br>
Let‚Äôs take the classic example from the slide:

$$
\vec{v} = 2\vec{i} + 3\vec{j}
$$

where

$$
\vec{i} = \begin{bmatrix} 1 \\ 0 \end{bmatrix},
\qquad
\vec{j} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}.
$$


````python
# Define the unit vectors i and j
i = [1, 0]
j = [0, 1]

# Define the coefficients
alpha = 2
beta = 3

# Compute the linear combination manually
v = [alpha * i[0] + beta * j[0],
     alpha * i[1] + beta * j[1]]

print("v =", v)
````

We just constructed a new vector [2, 3] by scaling i and j. <br>
In 2D space, that‚Äôs a point 2 units across, 3 units up.

````python
# in numpy code
import numpy as np

i = np.array([1, 0])
j = np.array([0, 1])

alpha, beta = 2, 3

v = alpha * i + beta * j
print("v =", v)
````

üìò **Span**

The **span** of a set of vectors is the set of all possible **linear combinations** you can form from them.

$$
\text{span}\{\vec{v}_1, \vec{v}_2\}
= \{\, c_1 \vec{v}_1 + c_2 \vec{v}_2 \;\mid\; c_1, c_2 \in \mathbb{R} \,\}
$$

---

üí¨ **Example:**  
If you take all combinations of  

$$
\vec{i} = \begin{bmatrix} 1 \\ 0 \end{bmatrix},
\qquad
\vec{j} = \begin{bmatrix} 0 \\ 1 \end{bmatrix},
$$

you can get **every vector in** $\mathbb{R}^2\$ ‚Äî because they **span the entire 2D plane.**

````python
# Demonstrating the span of i and j in R^2
i = [1, 0]
j = [0, 1]

# Try all linear combinations: c1*i + c2*j  where c1, c2 ‚àà { -2, -1, 0, 1, 2 }
combinations = [
    [a * i[0] + b * j[0], a * i[1] + b * j[1]]
    for a in range(-2, 3)
    for b in range(-2, 3)
]

print(combinations)

# Computes all possible linear combinations of vectors i and j
# Explores the span using integer scalars from -2 to 2
# But only prints the coordinates (text-only)
````

````python
import numpy as np

i = np.array([1, 0])
j = np.array([0, 1])

# All linear combinations a*i + b*j for a, b ‚àà { -2, -1, 0, 1, 2 }
combinations = np.array([a*i + b*j for a in range(-2, 3) for b in range(-2, 3)])

print(combinations)
````

Note: Each row is a vector in the span of i and j

````python
import numpy as np
import matplotlib.pyplot as plt

# Basis vectors
i = np.array([1, 0])
j = np.array([0, 1])

# Generate span combinations (c1*i + c2*j for c1, c2 in [-2, 2])
points = []
for a in range(-2, 3):
    for b in range(-2, 3):
        points.append(a * i + b * j)

points = np.array(points)

# Plot
plt.figure(figsize=(5, 5))
plt.scatter(points[:, 0], points[:, 1])  # x and y coordinates

# Styling
plt.axhline(0, linewidth=1)  # x-axis
plt.axvline(0, linewidth=1)  # y-axis
plt.title("Span of i and j (standard basis of R^2)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Basis vectors
i = np.array([1, 0])
j = np.array([0, 1])

# Scalars (you can change this range to go bigger)
alphas = range(-2, 3)
betas = range(-2, 3)

plt.figure(figsize=(5,5))

for a in alphas:
    for b in betas:
        v = a * i + b * j  # linear combination
        # draw vector from origin to v
        plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)
        # label (a,b) next to the tip
        plt.text(v[0] + 0.1, v[1] + 0.1, f"({a},{b})")

plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.title("Span of i and j in ‚Ñù¬≤ (Vector Arrows)")
plt.axhline(0, linewidth=1)
plt.axvline(0, linewidth=1)
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")

plt.show()
````

### üìå Linear Independence

A set of vectors  

$$
\vec{v}_1, \; \vec{v}_2, \; \dots, \; \vec{v}_k
$$

is said to be **linearly independent** if the **only** solution to

$$
c_1 \vec{v}_1 + c_2 \vec{v}_2 + \cdots + c_k \vec{v}_k = \vec{0}
$$

is

$$
c_1 = c_2 = \cdots = c_k = 0.
$$

---

üí¨ **In plain words:**  
None of the vectors can be created by combining the others.  
If even **one** vector can be written from the rest ‚Üí the set is **linearly dependent**.

---

### ‚úÖ Example of **Linearly Independent** vectors

$$
\vec{v}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix},
\quad
\vec{v}_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}
$$

The only way to make

$$
c_1 \vec{v}_1 + c_2 \vec{v}_2 = \vec{0}
$$

is if

$$
c_1 = 0, \quad c_2 = 0.
$$

‚úîÔ∏è Therefore, these vectors are **linearly independent**.

---

### ‚ùå Example of **Linearly Dependent** vectors

$$
\vec{v}_1 = \begin{bmatrix} 1 \\ 2 \end{bmatrix},
\quad
\vec{v}_2 = \begin{bmatrix} 2 \\ 4 \end{bmatrix}
$$

Notice:

$$
\vec{v}_2 = 2 \vec{v}_1
$$

So we can write:

$$
2\vec{v}_1 - 1\vec{v}_2 = \vec{0}
$$

with **non-zero coefficients**.

‚úîÔ∏è Therefore, these vectors are **linearly dependent**.


### üîπ In simple words

If the **only** way to make

$$
c_1 \vec{v}_1 + c_2 \vec{v}_2 = \vec{0}
$$

is by choosing

$$
c_1 = 0 \quad \text{and} \quad c_2 = 0
$$

‚Üí then the vectors are **linearly independent**.

**Why?**  
Because that means no vector can be formed using the others ‚Äî they are truly different directions.

---

### üî∏ Opposite case

If there exists even **one** case where

$$
c_1 \ne 0 \quad \text{or} \quad c_2 \ne 0
$$

and still

$$
c_1 \vec{v}_1 + c_2 \vec{v}_2 = \vec{0}
$$

‚Üí then the vectors are **linearly dependent** (one can be made from the other).

````python
# Define two vectors as plain Python lists
v1 = [1, 2]
v2 = [2, 4]   # Notice v2 = 2*v1 ‚Üí linearly dependent

# Try to solve c1*v1 + c2*v2 = [0, 0]
# Let's just test a specific non-zero combination:
c1, c2 = 2, -1

result = [
    c1 * v1[0] + c2 * v2[0],
    c1 * v1[1] + c2 * v2[1]
]

print("Result:", result)  # ‚Üí [0, 0]
print("Non-zero coefficients found ‚Üí Vectors are linearly dependent!")
````

````python
# numpy code
import numpy as np

v1 = np.array([1, 2])
v2 = np.array([2, 4])

c1, c2 = 2, -1
result = c1 * v1 + c2 * v2

print("Result:", result)  # ‚Üí [0 0]
print("Since we found non-zero (c1, c2), vectors are linearly dependent.")
````

Because we found non-zero scalars c1 = 2 and c2 = -1 that still give the zero vector, the vectors are dependent.

This matches the definition of linear dependence:

If there exists any combination of scalars not all zero that gives the zero vector, the vectors are dependent.

````python
# Define two vectors as plain Python lists
v1 = [1, 2]
v2 = [3, 1]   # Not a multiple of v1 ‚Üí linearly independent

# Try to solve c1*v1 + c2*v2 = [0, 0]
# Pick arbitrary non-zero scalars to test
c1, c2 = 2, -1

result = [
    c1 * v1[0] + c2 * v2[0],
    c1 * v1[1] + c2 * v2[1]
]

print("Result:", result)  # ‚Üí NOT [0, 0]
print("No non-zero coefficients give zero ‚Üí Vectors are linearly independent!")
````

````python
import numpy as np

# Define vectors as NumPy arrays
v1 = np.array([1, 2])
v2 = np.array([3, 1])  # Not a multiple of v1 ‚Üí linearly independent

# Pick arbitrary non-zero scalars to test
c1, c2 = 2, -1

# Compute linear combination
result = c1 * v1 + c2 * v2

print("Result:", result)  # ‚Üí [-1  3]
print("No non-zero coefficients give zero ‚Üí Vectors are linearly independent!")
````

Even if we pick c1 = 2 and c2 = -1, the linear combination:<br>
This shows there is no non-zero combination that produces the zero vector ‚Üí independent.

# SECTION 5: Basis and Dimension

### üìå A Basis of a Vector Space

A basis of a vector space 

$$
V \subseteq \mathbb{R}^n
$$

is a set of **linearly independent vectors** that **span the entire space**.

In other words, the basis gives us the ‚Äúdirections‚Äù we can move in to reach **any point** in the space.

A set 

$$
\{\vec{v}_1, \vec{v}_2, \dots, \vec{v}_k\}
$$

is a basis of \(V\) if:

1. It **spans** \(V\)  
2. The vectors are **linearly independent**

---

### üå∏ Example

In 2D space $\mathbb{R}^2\$ :  

The vectors

$$
\vec{i} = 
\begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad
\vec{j} = 
\begin{bmatrix} 0 \\ 1 \end{bmatrix}
$$

form the **standard basis**.

i = unit vector along x-axis

j = unit vector along y-axis

Together, {i, j} form the standard basis of ‚Ñù¬≤.

---
<br>

Every vector 

$$
\vec{v} = 
\begin{bmatrix} v_1 \\ v_2 \end{bmatrix}
$$

can be written as:

$$
\vec{v} = v_1 \vec{i} + v_2 \vec{j}
$$


````python
# Standard basis vectors in R2
i = [1, 0]
j = [0, 1]

# Any vector in R2
v = [2, 3]

# Express v as a linear combination of i and j
v1, v2 = v[0], v[1]

print("v = {} * i + {} * j".format(v1, v2))
````

````python
import numpy as np

# Standard basis vectors in R2
i = np.array([1, 0])
j = np.array([0, 1])

# Any vector in R2
v = np.array([2, 3])

# Express v as a linear combination of i and j
v1, v2 = v[0], v[1]

print(f"v = {v1} * i + {v2} * j")
````

i and j are there to show the idea of decomposition into a basis.

Even though we don‚Äôt ‚Äúuse‚Äù them numerically in the calculation, they represent the directions along which the vector is scaled.

Conceptually:

2*i ‚Üí move 2 units along x-axis

3*j ‚Üí move 3 units along y-axis

Together: vector v = [2,3]

### üß© Change of Basis

Sometimes, we can describe the **same vector** from a different perspective ‚Äî using a **different basis**.

**Example** (from your lecturer‚Äôs slide):

Standard basis:

$$
\vec{i} = 
\begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad
\vec{j} = 
\begin{bmatrix} 0 \\ 1 \end{bmatrix}
$$

and a **new basis**:

$$
\vec{w} = 
\begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad
\vec{z} = 
\begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

Let:

$$
\vec{v} = 
\begin{bmatrix} 2 \\ 3 \end{bmatrix}
$$

We want to find coordinates $(v_1^*, v_2^*)$ such that: 


$$
\vec{v} = v_1^* \vec{w} + v_2^* \vec{z}
$$

````python
# Define the vector in old basis
v = [2, 3]

# Define the new basis vectors
w = [1, 0]
z = [1, 1]

# Solve c1*w + c2*z = v for c1 and c2
# We solve the system manually:
# v[0] = c1*w[0] + c2*z[0] --> 2 = c1*1 + c2*1 --> c1 + c2 = 2
# v[1] = c1*w[1] + c2*z[1] --> 3 = c1*0 + c2*1 --> 0*c1 + 1*c2 = 3 --> c2 = 3

c2 = v[1] / z[1]  # c2 = 3 / 1 = 3
c1 = v[0] - c2    # c1 + c2 = 2 --> c1 = 2 - 3 = -1

print("Coordinates of v in the new basis:", [c1, c2])
````

````python
import numpy as np

# Define old basis (standard)
v = np.array([2, 3])

# Define new basis
w = np.array([1, 0])
z = np.array([1, 1])

# Combine basis vectors as columns in a matrix
B = np.column_stack((w, z))

# Solve for the coordinates in the new basis
v_star = np.linalg.solve(B, v)
print("Coordinates of v in the new basis:", v_star)
````

Different basis is the perspective from which we
observe the same vector in different ways

### üßÆ Dimension

The **dimension** of a vector space is equal to the **number of vectors in its basis**.

- $\mathbb{R}^2$ ‚Üí 2D (needs 2 independent vectors)  
- $\mathbb{R}^3$ ‚Üí 3D (needs 3 independent vectors)  

üí¨ **In Machine Learning:**  
The dimension corresponds to the **number of features** in your dataset.

# üè° SECTION 6: Application ‚Äî Predicting House Prices Using Vectors & Change of Basis

### üß† Concept Recap

> ‚ÄúWe wish to predict the price of a house.‚Äù

We‚Äôre told each house is represented by **two features**:

- Number of bedrooms  
- Number of bathrooms  

That means each house is a **vector in** $\mathbb{R}^2$:

$$
\vec{h} = \begin{bmatrix} \text{bedrooms} \\ \text{bathrooms} \end{bmatrix}
$$

---

### üèóÔ∏è Standard Basis Representation

We can represent these features using the **standard basis vectors**:

$$
\vec{i} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad
\vec{j} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}
$$

- Each direction represents one feature axis:  
  - $\vec{i}$ ‚Üí bedrooms  
  - $\vec{j}$ ‚Üí bathrooms

So, for a house with **3 bedrooms** and **2 bathrooms**:

$$
\vec{h} = 3\vec{i} + 2\vec{j} = \begin{bmatrix} 3 \\ 2 \end{bmatrix}
$$

````python
import numpy as np

# Standard basis vectors
i = np.array([1, 0])
j = np.array([0, 1])

# House: 3 bedrooms, 2 bathrooms
h = 3 * i + 2 * j
print("House vector h =", h)
````

üí¨ Interpretation: <br>
The coordinates (3, 2) tell us: <br>
‚Äú3 steps in the bedrooms direction, 2 steps in the bathrooms direction.‚Äù <br>

### üîÑ Change of Basis: Seeing Data Differently

Your lecturer introduced a **new basis** to get a better perspective:

$$
\vec{i}^* = \begin{bmatrix} 1 \\ 1 \end{bmatrix}, \quad
\vec{j}^* = \begin{bmatrix} 1 \\ -1 \end{bmatrix}
$$

üí¨ **Intuition:**

- $\vec{i}^*$: total number of rooms (bedrooms + bathrooms)  
- $\vec{j}^*$: difference between bedrooms and bathrooms  

So now, instead of thinking ‚Äúbedrooms‚Äù and ‚Äúbathrooms,‚Äù  
we think in terms of **total rooms** and **difference**.  
This is a **feature transformation** ‚Äî something ML models do all the time!  

````python
import numpy as np

# Old representation (bedrooms, bathrooms)
h = np.array([3, 2])

# New basis vectors
i_star = np.array([1, 1])
j_star = np.array([1, -1])

# Combine into a basis matrix
B_star = np.column_stack((i_star, j_star))

# Solve for new coordinates (v* = B*^-1 * h)
h_star = np.linalg.solve(B_star, h)
print("Coordinates of h in new basis (total, difference):", h_star)
````

### üß© Interpretation

The coordinates of a house in the new basis:

$$
\vec{h}^* = \begin{bmatrix} 2.5 \\ 0.5 \end{bmatrix}
$$

This means:

- The house has an overall **room intensity** of 2.5 (total rooms).  
- And a **difference** (imbalance) of 0.5 between bedrooms and bathrooms.

üí¨ **In ML terms:**  
This change of basis is similar to **feature engineering** ‚Äî transforming features into a space where patterns (like total room count) are clearer for the model to learn.


### üí° ML Connection

In machine learning:

Each basis vector = a feature direction.

Each data point = a vector.

Each model (like linear regression) = finds a new basis or direction that best fits/predicts outputs.

Even Principal Component Analysis (PCA) ‚Äî a dimensionality reduction algorithm ‚Äî is essentially a change of basis using eigenvectors.

Every data point in machine learning is a vector.
The way we choose to represent or transform these vectors ‚Äî through basis changes, linear combinations, and vector spaces ‚Äî determines how clearly patterns can be seen.
So, when we preprocess data, scale features, or perform PCA, we‚Äôre really just doing linear algebra magic ‚Äî giving our data a new pair of eyes.
