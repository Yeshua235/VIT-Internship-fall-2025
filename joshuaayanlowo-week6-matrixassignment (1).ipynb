## **Task 1 — Matrix Inverse & Determinant (4 marks)**
Given matrix \(A\):

$
A = \begin{bmatrix} 4 & 7 \\ 2 & 6 \end{bmatrix}
$

- **1A (2 marks):** Compute the inverse of \(A\) (show method or code).
- **1B (1 mark):** Compute \(\det(A)\).
- **1C (1 mark):** Verify that \(A A^{-1} = I\) (show code/verification).


```python
# Task 1: Implement here (students)
# You may use pure Python or NumPy. Example placeholders:
import numpy as np
from pprint import pp

A = [[4,7],[2,6]]
A = np.array(A, dtype=int)
# 1a
A_inv = np.linalg.inv(A)

# 1b
det_A = np.linalg.det(A)

# 1c
confirmation = A@A_inv

print('Given A = ')
pp(np.array(A))
print('1A.\nA_inv = ')
pp(A_inv)
print('1B.\ndeterminant of A = ', det_A)
print('\n1C.\nAA^-1 = ')
pp(confirmation)
```

## **Task 2 — Solving Linear Systems (3 marks)**
Solve the linear system:
\[
3x + y - z = 1 \\
2x + 4y + z = 2 \\
-x + 2y + 5z = 3
\]
- **2A (2 marks):** Find \(x, y, z\).
- **2B (1 mark):** Verify your solution by substituting back into the system.


```python
# Task 2: Implement here
A = [[3,1,-1],[2,4,1],[-1,2,5]]
b = [1,2,3]
# --- your code ---
A = np.array(A, dtype=int)
b = np.array(b, dtype=int)

# 2a
solution = np.linalg.solve(A, b)
x, y, z = round(solution[0], 2), round(solution[1], 2), round(solution[2], 2)
print('2A.')
print(f'x = {x}\ny = {y}\nz = {z}')

# 2b
equ_1 = 3*x + y - z
equ_2 = 2*x + 4*y + z
equ_3 = -x + 2*y + 5*z
print('\n2B.')
print('3x + y - z = 1; ', round(equ_1, 1) == 1)
print('2x + 4y + z = 2; ', round(equ_2, 1) == 2)
print('-x + 2y + 5z = 3; ', round(equ_3, 1) == 3)
```

## **Task 3 — Linear Transformations (4 marks)**
Consider the transformation matrix:
\[ T = \begin{bmatrix} 2 & 0.5 \\ 0 & 1 \end{bmatrix} \]
- **3A (1 mark):** Apply \(T\) to the vector \(v = [1,2]\).
- **3B (1 mark):** Plot the original vector and the transformed vector on the same axes.
- **3C (1 mark):** In one sentence, describe what the transformation does geometrically.
- **3D (1 mark):** Compute \(\det(T)\) and explain what it indicates about area scaling.


```python
# Task 3: Implement here
import numpy as np
T = np.array([[2,0.5],[0,1]])
v = np.array([1,2])
# --- your code ---
# 3a

v_transform = T@v
v_transform
print('Vector v = ', v)
print('Transformed v = ', v_transform)
```

```python
# 3b
import matplotlib.pyplot as plt

plt.plot([0, v[0]], [0, v[1]], color='r', label='before transformation')
plt.plot([0, v_transform[0]], [0, v_transform[1]], color='b', label='after transformation')
plt.axis('equal')
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.grid(True)
plt.legend()
plt.title('Linear Transformation in 2D')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show();
```

*3c.*

***Shearing***

```python
# 3d
det_T = np.linalg.det(T)
print('Det(T) = ', det_T)

print(f'The Area is scaled by a factor of {det_T}')
```

## **Task 4 — Geometric Transformation of a Shape (4 marks)**
Triangle vertices: \((0,0), (1,0), (0.5,1)\)
Apply the matrix:
\[ M = \begin{bmatrix} 1.5 & 0.5 \\ 0.2 & 1 \end{bmatrix} \]
- **4A (2 marks):** Plot the original triangle and the transformed triangle on the same axes.
- **4B (2 marks):** In two sentences, describe how the triangle changed (scale, shear, rotate, reflect etc.).


```python
# Task 4: Implement here
import numpy as np
triangle = np.array([[0,1,0.5,0],[0,0,1,0]])  # closed polygon (x coords; y coords)
M = np.array([[1.5,0.5],[0.2,1]])
# --- your code ---

transformed_triangle = M@triangle

# 4a
plt.figure(figsize=(10, 6))
# original triangle
plt.plot(triangle[0], triangle[1], color='blue', label='Before Transformation')
for i in range(3):    
    plt.text(triangle[0][i], triangle[1][i], f'{i}', fontsize=15)

# transformed triangle
plt.plot(transformed_triangle[0], transformed_triangle[1], color='red', label='After Transformation')
for i in range(3):    
    plt.text(transformed_triangle[0][i], transformed_triangle[1][i], f'{i}', fontsize=15)

plt.title('Linear Transformation of a Triangle')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend(loc='upper left')
plt.grid(True)
plt.show();
```

1. ***The sides are scaled***
2. ***The Triangle is rotated counterclockwisely.***

## **Task 5 — Change of Basis (3 marks)**
Let the basis matrix be:
\[ B = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}, \quad v = [3,2] \]
- **5A (1 mark):** Compute coordinates of \(v\) in the basis \(B\) (i.e. solve \(Bc = v\)).
- **5B (1 mark):** Convert the coordinates back to the standard basis.
- **5C (1 mark):** Verify correctness (show the reconstruction equals the original vector).


```python
# Task 5: Implement here
import numpy as np
B = np.array([[2,1],[1,2]])
v = np.array([3,2])
# --- your code ---

# 5a
c = (np.linalg.inv(B))@v
print("coordinates in new basis: ", c)

# 5b
v_standard = B@c
print("reconstruction of vector, v: ", v_standard)

# 5c
print('Is reconstruction correct: ', np.array_equal(v_standard, v))
```
